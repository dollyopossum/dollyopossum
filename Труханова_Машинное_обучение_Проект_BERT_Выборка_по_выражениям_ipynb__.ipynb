```python
#импорты
import pandas as pd
import numpy as np

import nltk

from nltk.tokenize import word_tokenize, wordpunct_tokenize

from nltk import download
download('punkt')

download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('russian')

#импорты
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

#Mystem
from pymystem3 import Mystem
mystem = Mystem()
```

```python
#создание датафрейма
df_chat = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Chat All ML.txt', header=None)
df_chat.columns = ['post']
```

```python
#вывод датафрейма
df_chat
```

```python
#исключение пустых ячеек(строк)
df_chat=df_chat.dropna()
```

```python
#создание списка из колонки датафрейма
chat_text = df_chat['post'].tolist()
print(len(chat_text))
```

```python
#лемматизация
chat_text_lemmatized = []
for i in range(len(chat_text)):
    chat_text_lemmatized.append(mystem.lemmatize(chat_text[i]))
print(len(chat_text_lemmatized))
```

```python
#очистка от пунктуации
chat_text_clean = []
for post in chat_text_lemmatized:
    post_text = []
    for word in post:
        if word[0].isalpha():
            post_text.append(word)
    chat_text_clean.append(post_text)
print(len(chat_text_clean))
```

```python
#обратное объединение списка списков лемм в список предложений (постов)
chat_text_list = []
for post in chat_text_clean:
    post_text = " ".join(post)
    chat_text_list.append(post_text)
print(len(chat_text_list))
```

```python
#очищение списка постов от пустых элементов
chat_text_list = list(filter(None, chat_text_list))
print(len(chat_text_list))
```

```python
#загрузка списка фраз для выборки
phrases_list = []
with open ('/content/drive/MyDrive/Colab Notebooks/phrases.txt', 'r', encoding='utf-8') as f:
  for i in f:
    if i[0].isalpha():
      phrases_list.append(mystem.lemmatize(i))
```

```python
#очистка от пунктуации
phrases_text = []
for phrase in phrases_list:
    phrase_text = []
    for word in phrase:
        if word[0].isalpha():
            phrase_text.append(word)
    phrases_text.append(phrase_text)
print(phrases_text)
```

```python
#обратное объединение списка списков в один список
phrases = []
for phrase in phrases_text:
    phrase_text = " ".join(phrase)
    phrases.append(phrase_text)
print(phrases)
```

```python
# применение фильтра, создание выборки
chat_text_list_phrases = []
for post in chat_text_list:
  for phrase in phrases:
    if post.find(phrase) != -1:
      chat_text_list_phrases.append(post)
      break
print(len(chat_text_list_phrases))
```

```python
#дополнительный список стопслов
add_stop = []
with open ('/content/drive/MyDrive/Colab Notebooks/Stop 2.txt', 'r', encoding='utf-8') as f:
  for i in f:
    add_stop.append(i.strip())
print(add_stop)
```

```python
#общий список стопслов
stop_words_all = stop_words + add_stop
stop_words_all = list(set(stop_words_all))
print(stop_words_all)
```

```python
#разбивка леммы
post_list_all = []
for post in chat_text_list_phrases:
  post_list = post.split()
  post_list_all.append(post_list)
```

```python
#очистка выборки от стопслов
chat_text_list_phrases_clean = []
for phrase in post_list_all:
    phrase_text = []
    for word in phrase:
        if word not in stop_words_all:
            phrase_text.append(word)
    chat_text_list_phrases_clean.append(phrase_text)
print(len(chat_text_list_phrases_clean))
```

```python
#очищение выборки от пустых элементов
chat_text_list_phrases_clean = list(filter(None, chat_text_list_phrases_clean))
print(len(chat_text_list_phrases_clean))
```

```python
#обратное объединение списка списков лемм в список предложений (постов)
chat_phrases_clean_list = []
for post in chat_text_list_phrases_clean:
    post_text = " ".join(post)
    chat_phrases_clean_list.append(post_text)
print(len(chat_phrases_clean_list))
```

# Тематическое моделирование с использованием эмбеддингов BERT

```python
#дополнительные импорты для работы с BERT
!pip install transformers
!pip install pyyaml

import torch
import json

from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from torch.utils.data import Dataset, DataLoader
from transformers import BertModel, BertTokenizer
```

```python
#подгрузка токенизатора и модели, создание для них сокращенных переменных
tokenizer = BertTokenizer.from_pretrained('DeepPavlov/rubert-base-cased')
model = BertModel.from_pretrained('DeepPavlov/rubert-base-cased', output_hidden_states = True)
```

```python
#создание двух корпусов из списка постов (со стопсловами и без стопслов)
df_chat_1 = pd.DataFrame(chat_text_list_phrases, columns=['post'])
corpus_1 = df_chat_1['post'].to_numpy()

df_chat_2 = pd.DataFrame(chat_phrases_clean_list, columns=['post'])
corpus_2 = df_chat_2['post'].to_numpy()
```

```python
#cоздание класса CustomDataset на основе импортированного класса Dataset с работы переопределением методов __len__ и __getitem__
#добавление метода tokenize, внутри которого к тексту будет применяться инициализированный ранее tokenizer
#установка максимальной длины токенизированного текста равной 512 токенам
#установка размера batch, в данном случае равного 50

class CustomDataset(Dataset):

    def __init__(self, X):
        self.text = X

    def tokenize(self, text):
        return tokenizer(text, return_tensors='pt', padding='max_length', truncation=True, max_length=512)

    def __len__(self):
        return self.text.shape[0]

    def __getitem__(self, index):
        output = self.text[index]
        output = self.tokenize(output)
        return {k: v.reshape(-1) for k, v in output.items()}

eval_ds_1 = CustomDataset(corpus_1)
eval_dataloader_1 = DataLoader(eval_ds_1, batch_size=50)

eval_ds_2 = CustomDataset(corpus_2)
eval_dataloader_2 = DataLoader(eval_ds_2, batch_size=50)
```

```python
#получение эмбеддингов
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output['last_hidden_state']
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    sum_embeddings = torch.sum(token_embeddings * input_mask_expanded, 1)
    sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)
    return sum_embeddings / sum_mask
```

```python
#перенос модели на графический ускоритель, если он доступен
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
print(device)
model.to(device)
model.eval()
```

```python
#перевод модели в состояние валидации и отключение подсчета градиента (корпус со стопсловами)
embeddings_1 = torch.Tensor().to(device)

with torch.no_grad():
    for n_batch, batch in enumerate(tqdm(eval_dataloader_1)):
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        embeddings_1 = torch.cat([embeddings_1, mean_pooling(outputs, batch['attention_mask'])])
    embeddings_1 = embeddings_1.cpu().numpy()
```

```python
#перевод модели в состояние валидации и отключение подсчета градиента (корпус без стопслов)
embeddings_2 = torch.Tensor().to(device)

with torch.no_grad():
    for n_batch, batch in enumerate(tqdm(eval_dataloader_2)):
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        embeddings_2 = torch.cat([embeddings_2, mean_pooling(outputs, batch['attention_mask'])])
    embeddings_2 = embeddings_2.cpu().numpy()
```

```python
#уменьшение размерности эмбеддингов с 768 до 15 с помощью метода главных компонент для ускорения расчётов
pca = PCA(n_components=15, random_state=42)
emb_15d_1 = pca.fit_transform(embeddings_1)
emb_15d_2 = pca.fit_transform(embeddings_2)
```

```python
print(len(emb_15d_1))
print(len(emb_15d_2))
```

```python
#запуск алгоритма кластеризации, пороговое расстояние (distance_threshold) равно 0,6,
#параметр linkage со значением 'average' означает, что координата считается, как среднее координат всех наблюдений, принадлежащих к этому кластеру,
#параметр affinity со значением 'cosine' означает, что для вычисления расстояния между наблюдениями используется косинусное расстояние.

clustering_bert_1 = AgglomerativeClustering(n_clusters=None, distance_threshold=0.6, affinity='cosine', linkage='average').fit(emb_15d_1)
clustering_bert_2 = AgglomerativeClustering(n_clusters=None, distance_threshold=0.6, affinity='cosine', linkage='average').fit(emb_15d_2)
```

```python
#уменьшение размерности до двух для визуализации результата на графике
pca = PCA(n_components=2, random_state=42)
emb_2d_1 = pd.DataFrame(pca.fit_transform(embeddings_1), columns=['x1', 'x2'])
emb_2d_1['label'] = clustering_bert_1.labels_
emb_2d_1['label'].nunique()
```

```python
emb_2d_2 = pd.DataFrame(pca.fit_transform(embeddings_2), columns=['x1', 'x2'])
emb_2d_2['label'] = clustering_bert_2.labels_
emb_2d_2['label'].nunique()
```
